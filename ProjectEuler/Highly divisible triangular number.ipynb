```python
def is_prime(n):
    sqrt = (math.sqrt(n))
    print sqrt, int(sqrt)
    if n == 2:
        return "Prime"
    
    for fac in xrange(2, int(sqrt)+1):
        if n%fac == 0:
            print fac
            return "Composite"
    return "Prime"
```
